## Part 1

```python
import pandas as pd
report = pd.read_csv('input.csv', names=['depth'])
```

```python
report['next_value']=report['depth'].shift(-1)
report['change'] = report['depth'] < report['next_value'] 
report[:20]
```

```python
sum(report.change)
```

## Part 2

```python
report = pd.read_csv('input.csv', names=['depth'])
report
```

```python
df = report.rolling(window=3).sum().dropna()
```

```python
df['next_value']=df['depth'].shift(-1)
df['change'] = df['depth'] < df['next_value'] 
df[:20]
```

```python
sum(df.change)
```
